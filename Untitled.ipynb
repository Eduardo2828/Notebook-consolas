```python
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# 1️⃣ Cargar el archivo CSV
# ===============================
df = pd.read_csv("consoles3.csv")

# ===============================
# 2️⃣ Limpiar y preparar los datos
# ===============================
df = df.dropna(subset=["Año_de_Lanzamiento", "Ventas_Globales"])
df["Año_de_Lanzamiento"] = df["Año_de_Lanzamiento"].astype(int)

# ===============================
# 3️⃣ Filtrar por el año 2010
# ===============================
df_2010 = df[df["Año_de_Lanzamiento"] == 2010]

# ===============================
# 4️⃣ Agrupar por consola/plataforma
# ===============================
ventas_por_consola_2010 = (
    df_2010.groupby("Plataforma")["Ventas_Globales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

print("Consolas más usadas (más ventas) en 2010:")
print(ventas_por_consola_2010)

# ===============================
# 5️⃣ Gráfico tipo “queso”
# ===============================
plt.figure(figsize=(8, 8))
plt.pie(
    ventas_por_consola_2010["Ventas_Globales"],
    labels=ventas_por_consola_2010["Plataforma"],
    autopct="%1.1f%%",
    startangle=90,
    shadow=True,
    explode=[0.05]*len(ventas_por_consola_2010)
)
plt.title("Consolas más usadas en 2010 (por ventas globales)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# 1️⃣ Cargar el archivo CSV
# ===============================
df = pd.read_csv("consoles3.csv")

# ===============================
# 2️⃣ Limpiar y preparar los datos
# ===============================
df = df.dropna(subset=["Género", "Ventas_Globales"])

# ===============================
# 3️⃣ Agrupar por género y sumar ventas globales
# ===============================
ventas_por_genero = (
    df.groupby("Género")["Ventas_Globales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

print("🎮 Ventas globales por género:")
print(ventas_por_genero)

# ===============================
# 4️⃣ Crear gráfico de barras
# ===============================
plt.figure(figsize=(10, 6))
plt.bar(
    ventas_por_genero["Género"],
    ventas_por_genero["Ventas_Globales"]
)
plt.title("Ventas Globales Totales por Género de Videojuego")
plt.xlabel("Género")
plt.ylabel("Ventas Globales (en millones)")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# 1️⃣ Cargar el archivo CSV
# ===============================
df = pd.read_csv("consoles3.csv")

# ===============================
# 2️⃣ Limpiar y preparar los datos
# ===============================
df = df.dropna(subset=["Año_de_Lanzamiento", "Puntuación_Crítica", "Puntuación_Usuarios"])
df["Año_de_Lanzamiento"] = df["Año_de_Lanzamiento"].astype(int)

# ===============================
# 3️⃣ Encontrar el juego con mejor crítica y usuarios de cada año
# ===============================
mejor_critica = df.loc[df.groupby("Año_de_Lanzamiento")["Puntuación_Crítica"].idxmax(),
                       ["Año_de_Lanzamiento", "Título", "Puntuación_Crítica"]].sort_values("Año_de_Lanzamiento")

mejor_usuarios = df.loc[df.groupby("Año_de_Lanzamiento")["Puntuación_Usuarios"].idxmax(),
                        ["Año_de_Lanzamiento", "Título", "Puntuación_Usuarios"]].sort_values("Año_de_Lanzamiento")

# ===============================
# 4️⃣ Graficar la evolución en líneas
# ===============================
plt.figure(figsize=(10, 6))
plt.plot(mejor_critica["Año_de_Lanzamiento"], mejor_critica["Puntuación_Crítica"],
         marker="o", label="Mejor Puntuación Crítica")
plt.plot(mejor_usuarios["Año_de_Lanzamiento"], mejor_usuarios["Puntuación_Usuarios"],
         marker="s", label="Mejor Puntuación Usuarios")

plt.title("Evolución de las Mejores Puntuaciones de Crítica y Usuarios por Año")
plt.xlabel("Año de Lanzamiento")
plt.ylabel("Puntuación")
plt.legend()
plt.grid(True)
plt.show()
```
