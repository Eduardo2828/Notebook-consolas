```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('consoles3.csv')

# Contar cuántos hay de cada valor en la columna "Género"
conteo_genero = df['Género'].value_counts()

# Mostrar el resultado
print(conteo_genero)
```

```python
conteo_genero_df = conteo_genero.reset_index()
conteo_genero_df.columns = ['Género', 'Cantidad']
print(conteo_genero_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el CSV
df = pd.read_csv('consoles3.csv')

# Contar los valores de la columna "Género"
conteo_genero = df['Género'].value_counts()

# Graficar un gráfico de barras
conteo_genero.plot(kind='bar')

# Añadir título y etiquetas
plt.title('Cantidad por Género')
plt.xlabel('Género')
plt.ylabel('Cantidad')

# Mostrar el gráfico
plt.show()
```
